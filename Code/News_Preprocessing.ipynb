```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Job_News.csv')
data.head()
```

```python
data.rename(columns = {col: col.strip().replace(" ", '') for col in data.columns}, inplace = True)
```

```python
data.isna().sum()
```

```python
data.drop("Source", inplace = True, axis = 1)
```

```python
data.head()
```

```python
sns.catplot(data = data, x = 'Source_Name', kind = "count", height=10, aspect=2)
plt.xticks(rotation = 45);
```

```python
order = data['Source_Name'].value_counts().index
sns.catplot(data=data, x='Source_Name', kind="count", order=order[:10], height=10, aspect=2)
plt.xticks(rotation=45)
plt.show()
```
